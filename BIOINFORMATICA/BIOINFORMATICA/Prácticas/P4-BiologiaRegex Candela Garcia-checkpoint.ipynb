##### 4.- Escribe una función que cuente el porcentaje de A de una secuencia

```python
def porcentaje_A(secuencia):
    total = len(secuencia)
    count_A = secuencia.count("A")
    return (count_A / total) * 100
```

```python
print(porcentaje_A("AATTGC"))  
```

##### ¿Puedes hacer que devuelva dos decimales? ¿Y que lidie con mayúsculas y minúsculas?

```python
def porcentaje_A(secuencia):
    secuencia = secuencia.upper()  # convierte todo a mayúsculas
    total = len(secuencia)
    count_A = secuencia.count("A")
    return round((count_A / total) * 100, 2)
```

```python
print(porcentaje_A("AaTtGC"))
```

##### ¿Y que devuelva el número de decimales que yo quiero?

```python
def porcentaje_A(secuencia, decimales=2):
    secuencia = secuencia.upper()
    total = len(secuencia)
    count_A = secuencia.count("A")
    return round((count_A / total) * 100, decimales)
```

```python
print(porcentaje_A("AaTtGC", 3))
print(porcentaje_A("AaTtGC", 1))
```

##### ¿Qué pasa si no ponemos un argumento? ¿Sabrías hacer una función encapsulada?

```python
def porcentaje_A_encapsulada(secuencia=None, decimales=2):
    if secuencia is None:
        secuencia = input("Introduce una secuencia: ")
    secuencia = secuencia.upper()
    total = len(secuencia)
    count_A = secuencia.count("A")
    return round((count_A / total) * 100, decimales)
```

```python
print(porcentaje_A_encapsulada("TtAaGG"))
```

```python
print(porcentaje_A_encapsulada())
```

#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re
```

```python
accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']
```

```python
print("1.", [a for a in accs if re.search(r'5', a)])
```

```python
print("2.", [a for a in accs if re.search(r'[de]', a, re.I)])
```

```python
print("3.", [a for a in accs if re.search(r'd.*e', a, re.I)])
```

```python
print("4.", [a for a in accs if re.search(r'd.e', a, re.I)])
```

```python
print("5.", [a for a in accs if re.search(r'd', a, re.I) and re.search(r'e', a, re.I)])
```

```python
print("6.", [a for a in accs if re.search(r'^[xy]', a, re.I)])
```

```python
print("7.", [a for a in accs if re.search(r'^[xy].*e$', a, re.I)])
```

```python
print("8.", [a for a in accs if re.search(r'\d{3,}', a)])
```

```python
print("9.", [a for a in accs if re.search(r'd[arp]$', a, re.I)])
```
