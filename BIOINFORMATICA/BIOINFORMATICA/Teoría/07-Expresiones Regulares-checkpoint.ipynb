# EXPRESIONES REGULARES

El módulo re contiene los métodos necesarios para analizar y buscar expresiones regulares en Python.

**import re**

**re.search(pattern,string)**

La función usa dos argumentos, el patrón y la cadena dónde hay que buscarlo


Para indicar a Python que ignore los caracteres especiales como *\t* o *\n* se puede usar *r* fuera de las comillas: **print(r”\t\n”)**

El resultado que devuelve **re.search()** es booleano, true/false

```python
import re
```

```python
# re.search(pattern,string)
# r se usa para que Python ignore caracteres especiales
```

```python
dna = "ATCGCGAATTGAFCAC"
if re.search(r"GAZ", dna):
    print("restriction enzyme site found!")
else:
    print("Not site found!")
```

```python
# Alternaciones
dna = "ATCGCGAATTCAC"
if re.search(r"ATT", dna) or re.search(r"CGAA", dna):
    print("restriction enzyme site found!")
```

```python
dna = "ATCGCGAATTCAC"
if re.search(r"GC(G|C)AATT", dna):
    print("restriction enzyme site found!")
```

```python
# Grupos de caracteres
dna = "ATCGCGGCAATTCAC"
if re.search(r"GC(A|T|G|C)GC", dna):
    print("restriction enzyme site found!")
```

```python
dna = "ATCGCGGCAATTCAC"
if re.search(r"GC[ATGC]GC", dna):
    print("restriction enzyme site found!")	
```

```python
# Negaciones
dna = "ATCGCGAATTCAC"
if re.search(r"GC[^N]", dna): #todo menos la N
    print("restriction enzyme site found!")
```

```python
# Extración de parte del patrón en una variable
dna = "ATGACGTACGTACGACTG"
# guarda los patrones en la variable m
m = re.search(r"GA[ATGC]{3}AC",dna) #la llave 3 es una llave de repetición: uno de esos 3 veces
m_pattern = m.group() #esta variable m engloba todos los patrones que he guardado, los puedo ver con group
print(m_pattern)
```

```python
# Extrae un más de un bit del patrón
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("entire match: " + m.group()) # no pongo nada me imprime todo el patron
print("first bit: " + m.group(1))#pongo 1 imprime el primer patron
print("second bit: " + m.group(2)) #pongo 2 imprime el segundo patron
```

```python
# Algunos métodos
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("input string: " + dna)
print("entire match: " + m.group())
```

```python
# Start(): el inicio, en que posicion empieza el patron: ej sitio de restriccion de una enzima 
print("start: " + str(m.start())) 
```

```python
# End()
print("end: " + str(m.end()))
```

```python
# Split(): divide el caracter de entradad por la C
dna = "ACTNGCATRGCTACGTYACGATSCGAWTCG"
runs = re.split(r"C", dna)
print(runs)	
```

```python
# Extraer  usando group()
dna = "ATGACGTACGTACGACTG"
m = re.search(r"GA([ATGC]{3})AC([ATGC]{2})AC", dna)
print("input string: " + dna)
print("entire match: " + m.group())
print("first bit: " + m.group(1))
print("second bit: " + m.group(2))
```

```python
# Finditer() detecta las bases erróneas: encontramos el patron y donde empieza, tmb interesa para enzimas de restriccion
dna = "CGCTCNTAGATGCGCRATGACTGCAYTGC" 

matches = re.finditer(r"[^ATGC]", dna) 
for m in matches: 
    base = m.group() 
    pos  = m.start() 
    print(base + " found at position " + str(pos))
```

```python
dna = "CGCTCNTAGATGCGCRATGACTGCAYTGC" 

matches = re.finditer(r"[^ATGC]", dna) 

for i in matches:
    match = i.group()
    pos = i.start()
    print(f"La base {match} está en la posición {pos}")
```

```python
# Findall(): cuando tenemos patrones que se van a repetir varias veces. intersa para sacar los patrones de corte de una enzima de restriccion y de ahi sacar las posiciones
dna = "CTGCATTATATCGTACGAAATTATACGCGCG" 
result = re.findall(r"[AT]{6,}", dna) 
print(result)
```
