#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
def digestion_restriccion(secuencia, enzimas):# Definela función digestión ADN con enzs restriccion mediante dos parámetros:
    # secuencia: la cadena de ADN a digerir
    # enzimas: introduce el diccionario enzR dentro del parámetro enzimas, para que estén dentro de la función.
    seq = secuencia.replace("\n", "").replace(" ", "").upper() # Permite limpiar la secuencia:
    # secuencia.replace("\n", ""): elimina todos los saltos de línea (\n).
    # .replace(" ", ""): elimina todos los espacios.
    # .upper(): convierte la secuencia en mayúsculas, para que todas tengan el mismo patrón (FASTA a veces trae minúsculas)
    # Creamos estructuras para guardar los resultados: 
    cortes = {}        # Enzimas que encuentran un sitio y cortan.
    no_cortan = []     # Enzimas  que no aparecen en la secuencia y no se cortan.
    fragmentos = []    # Fragmentos producidos tras la digestión.

    # Buscamos los sitios de corte de cada enzima:
    for nombre, patron in enzimas.items(): # Recorre cada enzima y su patrón
        # Determinamos la posición de corte con: 
        corte_idx = patron.index('*') # index('*'): devuelve la posición exacta del corte dentro del patrón.
        sitio = patron.replace('*', '') # replace('*', '') elimina el asterisco para obtener el sitio de reconocimiento real.
        
    # Buscamos todas las apariciones del sitio en la secuencia.
        posiciones = [] # Crea una lista vacía donde se guardarán todas las posiciones de corte.
        start = 0 # Indica desde qué posición de la secuencia queremos empezar a buscar.
        while True: # Crea un bucle infinito porque la condición True es siempre verdadera, dado que no conocemos cuántos sitios de corte tiene la enzima.
            # El bucle continúa buscando sitios de reconocimiento hasta que find ya no encuentra más y se interrumpe con break.
            pos = seq.find(sitio, start) # Busca la siguiente aparición del sitio de reconocimiento en la secuencia:
            # seq.find(...): devuelve la posición si esta existe, y si no existe devuelve un -1.
            if pos == -1: # Comprueba el valor que devolvió find, si no se ha encontrado el sitio de reconocimiento.
                break # Si find devolvió -1, significa que ya no hay más sitios. Detiene el bucle.
            posiciones.append(pos + corte_idx)  # Si se encuentra el sitio de corte:
            # pos: posición donde empieza el sitio.
            # corte_idx: posición del * dentro del patrón
            start = pos + 1 # Actualiza la posición desde donde se está buscando, para que find avance a la siguiente posición y no encuentre siempre el mismo sitio.
        if posiciones: # Si la lista posiciones, que hemos creado anteriormente, no está vacía, añade a cortes la enzima y las posiciones donde corta.
            cortes[nombre] = posiciones
        else: # Si no se encontró ningún sitio añade el nombre de la enzima a la lista no_cortan.
            no_cortan.append(nombre)
            
    # Para obtener todas las posiciones de corte de las enzimas:
    posiciones_corte = sorted([p for lista in cortes.values() for p in lista])
    # cortes.values(): obtiene todas las listas de posiciones de cada enzima.
    # [p for lista in cortes.values() for p in lista]: convierte todas esas listas en una sola lista con todas las posiciones.
    # sorted(...): ordena de menor a mayor
    
    # Para crear los fragmentos reales del ADN    
    inicio = 0 # Indica que el primer fragmento empieza al inicio de la secuencia, posición 0.
    for pos in posiciones_corte: # Recoore cada posición de corte y crea fragmentos: 
        fragmentos.append(seq[inicio:pos])
        # seq[inicio:pos]: extrae la parte del ADN desde el último corte hasta el nuevo corte.
        # fragmentos.append(...): añade ese fragmento a la lista de fragmentos.
        inicio = pos # Actualiza la posición del último corte.
    fragmentos.append(seq[inicio:])  # Después del último corte, queda un fragmento final que va hasta el final de la secuencia.

     # Mostramos los resultados.
    print("Enzimas que cortan y posiciones:") # Texto que aparece en pantalla.
    for enzR, pos in cortes.items(): # cortes.items(): devuelve pares (nombre_enzima, lista_de_posiciones).
        print(f" - {enzimas}: {pos}") # El bucle recorre cada enzima que sí corta la secuencia, mostrando su nombre y posición.
    
    if no_cortan: # Si la lista no_cortan tiene elementos, los imprime separados por comas.
        print("\nEnzimas que no cortan:", ", ".join(no_cortan))
    else: # Si no_cortan está vacía → todas las enzimas cortan al menos una vez.
        print("\nTodas las enzimas tienen al menos un sitio de corte.")

    print(f"\nNúmero total de fragmentos: {len(fragmentos)}") # Muestra cuántos fragmentos se generaron tras aplicar todos los cortes.
    print("Tamaños de los primeros 10 fragmentos:", [len(f) for f in fragmentos[:10]]) # Muestra el tamaño de los primeros 10 fragmentos.

    # Devolver los resultados
    return cortes, no_cortan, fragmentos
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
# Creamos un diccionario que traduzca cada codón de ADN (triplete de bases) al aminoácido correspondiente. 
# Esto te permitirá transformar cada ORF en su proteína correspondiente.
codon_table = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
    'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W'
}
# Definimos la función principal. Recibe como entrada la secuencia de ADN y, a partir de ella, obtendrá todos los ORFs posibles.
def analizar_ORFs(secuencia):
    # Define la función complementaria, que recibe una secuencia de ADN.
    def complementaria(seq):
        complement = {'A':'T','T':'A','C':'G','G':'C'} # Crea un diccionario con las reglas de complementariedad del ADN.
        return ''.join([complement[base] for base in seq])
        # [complement[base] for base in seq]: recorre cada base de la secuencia original, la sustituye por la base complementaria utilizando el diccionario y produce una lista.
        # ''.join(...): Une todos los elementos de la lista en una cadena continua sin espacios.
    # Define la función traducir, que recibe una secuencia de ADN y la convertirá en una proteína.
    def traducir(seq):
        prot ='' # Inicia una cadena vacía donde se irá construyendo la proteína final aminoácido por aminoácido.
        for i in range(0, len(seq)-2, 3): # Este bucle recorre la secuencia de tres en tres nucleótidos.
            codon = seq[i:i+3] # Extrae un codón (3 bases) desde la posición i hasta i+3.
            prot += codon_table.get(codon, 'X')  # Traduce el codón a su aminoácido correspondiente:
            # codon_table: busca el codón.
            # Si existe, devuelve su aminoácido.
            # Si no existe, pone "X" (aminoácido desconocido).
        return prot # Devuelve la secuencia proteica completa traducida.
        
    # Generamos las 6 traducciones posibles mediante: 
    strands = {'+': secuencia, '-': complementaria(secuencia)[::-1]}
    # secuencia: hebra directa (strand +).
    # complementaria(secuencia): genera la hebra complementaria.
    # [::-1]: invierte la cadena para obtener la hebra complementaria inversa, que es la hebra “–”.
    # De esta manera obtenemos 3 marcos de lectura en la hebra+ y otros 3 en la hebra -.
    orfs = [] # Creamos esta lista para almacenar todos los ORFs detectados.
    
    for strand_name, seq in strands.items(): # Este bucle analiza las 2 hebras de ADN. Recorre cada hebra y cada marco.
        # strand.items: devuelve pares (('+', secuencia_positiva)y ('-', secuencia_negativa))
        # strand_name: hebra '+' o '-'
        # seq: la secuencia correspondiente a esa hebra
        
        for frame in range(3): # Cada hebra tiene 3 formas distintas de agrupar los codones.
            # Para cada secuencia, se recorren los 3 marcos de lectura posibles:
            # frame = 0: empieza en nucleótido 0.
            # frame = 1: empieza en nucleótido 1.
            # frame = 2: empieza en nucleótido 2.
            # En total se generarán 6 posibles traducciones.
            prot = traducir(seq[frame:]) # Traduce el marco a proteína.
            
            # Para buscar un ORF en la proteína: 
            inicio = 0 # Recorre la proteína traducida desde el inicio.
            while inicio < len(prot): # Bucle que analiza la proteína aminoácido por aminoácido.
                if prot[inicio] == 'M': # Detecta el codón de inicio, si encuentra una M, inicia un ORF.
                    fin = inicio # Avanza hasta encontrar STOP, fin empieza en la misma posición donde apareció la M.
                    while fin < len(prot) and prot[fin] != '_': # Avanza continuamente mientras que no aparezca un stop.
                        # Cuando prot[fin] encuentra un codón STOP (_),termina el ORF.
                        fin += 1
                    orfs.append({'protein': prot[inicio:fin], 'strand': strand_name}) #Se almacena un diccionario con:
                    # protein: la secuencia proteica del ORF (desde la M hasta el STOP).
                    # strand: la hebra en la que se encontró ('+' o '-').
                    inicio = fin # Si encontró un STOP, sigue buscando a partir de ahí.
                else: 
                    inicio += 1 # Si NO es una M, avanza al siguiente aminoácido.
                    
    # Ordenamos los ORFs por longitud descendente
    orfs.sort(key=lambda x: len(x['protein']), reverse=True) # Creamos un diccionario.
    # .sort(...): ordena la lista.
    # key=lambda x: len(x['protein']): devuelve el tamaño de cada proteína de cada ORF (x= ORF).
    # everse=True: ordena de mayor a menor, primero los ORFs más largos.

    # Seleccionamos los 6 elementos más largos
    top6 = orfs[:6]

    # Mejor predicción:escogemos el ORF más largo
    mejor = top6[0] if top6 else None
    # mejor = top6[0]: guarda el ORF más largo de toda la secuencia, que es la mejor predicción de gen.
    # mejor = None: significa que no hay ORFs encontrados y por lo tanto, no existe ninguna predicción posible.

    # Calculamos % de aa aromáticos (F, Y, W) y la longitud relativa
    if mejor: # Verifica si se ha encontrado al menos un ORF válido.
        # Si no hay ninguno (mejor = None), no se calculan porcentajes ni longitudes.
        
        # Para contar aminoácidos aromáticos:
        arom = sum(mejor['protein'].count(a) for a in ['F','Y','W'])
        # ['F', 'Y', 'W']: Recorre la lista para buscar Fenilalanina, Tirosina y Triptófano, los tres aminoácidos aromáticos.
        # mejor['protein'].count(a): Cuenta cuántas veces aparece cada aminoácido aromático en la proteína.
        # sum: suma los resultados.
        
        # Para calcular el porcentaje: 
        porcentaje_aromaticos = (arom / len(mejor['protein'])) * 100 if len(mejor['protein'])>0 else 0
        # arom / len(mejor['protein']) → dividimos los aminoácidos aromáticos encontrados entre la longitud total de la proteína, para obtener la fracción de aa arom.
        # Multiplicamos *100 para convertir a porcentaje.
        
        # Calculamos la longitud relativa del ORF respecto al ADN
        longitud_relativa = (len(mejor['protein'])*3 / len(secuencia)) * 100
        # (len(mejor['protein']) * 3): calcula la longitud de la proteína en bases de ADN.
        # Como cada aminoácido procede de un codón, y cada codón son 3 bases de ADN, se multiplica por 3.
        # / len(secuencia): proporción del ORF respecto al total del ADN.

        # Los valores calculados (porcentaje de aromáticos y longitud relativa) se guardan en el diccionario del ORF.
        mejor['porcentaje_aromaticos'] = porcentaje_aromaticos
        mejor['longitud_relativa'] = longitud_relativa

    # Devolvemos los resultados:
    return orfs, top6, mejor
```

```python
# Llamamos a la función: analizar_ORFs.
orfs, top6, mejor_proteina = analizar_ORFs(secuencia)
# Ejecuta la función analizar_ORFs usando la secuencia de ADN.
# Guarda los tres resultados que devuelve la función en variables: orfs, top6 y mejor_proteina.

# Mostramos los 6 ORFs más largos.
print("6 ORFs más largos:")
for i, o in enumerate(top6, 1):
    print(f"{i}. Longitud: {len(o['protein'])} aa, Strand: {o['strand']}")
    # enumerate(top6, 1): numera los 6 ORFs empezando desde 1.
    # o es cada ORF dentro de top6.
    # o['protein'] es la secuencia proteica del ORF.
    # len(o['protein']) da su longitud en aminoácidos.
    # o['strand'] indica si está en la hebra + o -.

# Mostramos la mejor predicción.
if mejor_proteina: # Comprueba si existe un ORF más largo.
    print("\nMejor predicción (ORF más largo):") # Texto que queremos mostrar. 
    print("Longitud:", len(mejor_proteina['protein']), "aa") # Muestra cuántos aminoácidos tiene el ORF más largo.
    print("Strand:", mejor_proteina['strand']) # Indica si pertenece a la hebra + o -.
    print("Porcentaje de aminoácidos aromáticos:", round(mejor_proteina['porcentaje_aromaticos'],2), "%") # Muestra el porcentaje de aminoácidos aromáticos redondeando a dos decimales.
    print("Longitud relativa respecto al ADN:", round(mejor_proteina['longitud_relativa'],2), "%") # Muestra qué porcentaje de las bases del ADN ocupa ese ORF redondeando a dos decimales.
```
