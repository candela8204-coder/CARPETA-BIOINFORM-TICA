## JUEGO DE ADIVINANZAS

#### Escribe un programa que coja un número entero del 1 al 100, y los jugadores tienen que tratar de adivinar el número.

#### Las reglas son las siguientes:

1. Si un jugador escoje un número menor que 1 o mayor que 100, el programa dice "FUERA DE LÍMITES"
2. En el primer turno del jugador, si está en un rango de 10 del número dice "¡CALIENTE!". Si está fuera de ese rango dice "¡FRIO!"
3. En todas las rondas siguientes, si el jugador se acerca más que la vez anterior, dice "¡MAS CALIENTE!". Si está más lejos que la anterior vez dice "¡MAS FRIO!"
4. Cuando el jugador acierte el número, les dirá que han acertado y cuántos intentos les ha llevado





##### PISTAS
1. Trata obtener un número aleatorio entre el 1 y el 100
2. Haz una introducción de tu juego y explica las reglas
3. Crea una lista que almacene los intentos. 
4. Haz un bucle que evalúe un número. Prueba que funcione primero. While podría ir bien
5. Modifica ese bucle para que compare nuestro número con el intento. Si se acierta usa break, si no puedes usar continue y tienes que seguir preguntando al jugador por nuevos intentos

```python
import random #Import random para poder generar números aleatorios. 

def juego_adivinanzas():
    print("¡Bienvenido al juego de adivinanzas!")
    print("Tienes que adivinar un número entero entre 1 y 100.")
    print("Si introduces un número fuera de esos límites, te diré 'FUERA DE LÍMITES'.")
    print("En el primer intento, si estás a 10 o menos del número, te diré '¡CALIENTE!', sino '¡FRIO!'.")
    print("En intentos siguientes, te diré si estás '¡MAS CALIENTE!' o '¡MAS FRIO!' comparado con tu intento anterior.")
    print("¡Vamos a comenzar!")

    numero_objetivo = random.randint(1, 100) # Genera un número entero aleatorio entre 1 y 100
    intentos = [] #Se crea una lista vacía para almacenar todos los intentos del jugador

    while True: #Inicia un bucle infinito que seguirá pidiendo al jugador que ingrese un número hasta que acierte
        try:
            intento = int(input("Introduce tu intento: ")) 
            #int intenta convertir ese texto en un número entero.
            # input lee lo que el jugador escribe desde el teclado.
        except ValueError: #Si se escribe algo que no se puede convertir a entero,da error
            print("Por favor, introduce un número válido.")
            continue

        # Revisa que el numero esté en el rango permitido
        if intento < 1 or intento > 100:
            print("¡FUERA DE LÍMITES!")
            continue

        intentos.append(intento) #Se añade al intento a la lista de intentos

        # Si acierta
        # Se muestra cuántos intentos tomó (usando len(intentos) para contar elementos en la lista
        # Para finalizar el juego break termina el bucle
        if intento == numero_objetivo:
            print(f"¡Felicidades! Has acertado el número {numero_objetivo} en {len(intentos)} intentos.")
            break

        # Se ejecuta en el primer intento.
        # Abs calcula la diferencia entre el intento y el número objetivo
        # Si la diferencia es 10 o menos, el jugador está “cerca” y mostramos “¡CALIENTE!”, si no, “¡FRIO!”.
        if len(intentos) == 1:
            if abs(intento - numero_objetivo) <= 10:
                print("¡CALIENTE!")
            else:
                print("¡FRIO!")
        else:
            # Comparar con intento anterior
            distancia_actual = abs(intento - numero_objetivo) #distancia del intento actual al número objetivo.
            distancia_anterior = abs(intentos[-2] - numero_objetivo) #distancia del intento anterior (por eso usamos intentos[-2], que accede al penúltimo elemento).
            #Si la distancia actual es menor, el jugador se está acercando y mostramos “¡MAS CALIENTE!”.
            #Si es mayor o igual, está más lejos y mostramos “¡MAS FRIO!”.
            if distancia_actual < distancia_anterior:
                print("¡MAS CALIENTE!")
            else:
                print("¡MAS FRIO!")
```
